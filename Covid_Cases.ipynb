```
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


labels = ['U.P.', 'Maharashtra', 'Gujarat', 'New Delhi', 'Tamil Nadu']
total_positive = [20943, 152765, 30095, 77240, 74622 ]
total_recovered = [13583, 79815, 22030, 47091, 41357]

x = np.arange(len(labels))  
width = 0.35  

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, total_positive, width, label='Positive')
rects2 = ax.bar(x + width/2, total_recovered, width, label='Recovered')


ax.set_ylabel('Cases')
ax.set_title('Total Positive cases and Recovered till today')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```
